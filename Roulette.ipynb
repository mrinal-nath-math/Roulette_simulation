```python
"WITH HOUSE EDGE"
import numpy as np
import matplotlib.pyplot as plt
import random
def betting (money=100, goal = 1000, initial_bet = 1):
    position = np.zeros(goal + 1)
    position[0] = money
    bet = initial_bet
    idx = 0
    while money >= bet and idx < goal and money > 0  and money <= goal :
        spin = np. random. randint(-1, 36)
        idx += 1
        if spin % 2 == 0 and spin != 0:
          bet = initial_bet
          money += bet

        if spin % 2 !=0 and spin > 0:
            bet = initial_bet
            money -= bet
        if spin == 0 or spin == -1 :
          bet = initial_bet
          money -= bet

        position[idx] = money
    return position
position = betting(money = 100)
plt.plot(betting())
plt.yticks(np.arange(-10,200,10))
#plt.xticks(np.arange(0,1000,200))
plt.axhline (y=0, color='r', linestyle='dotted')
plt.axhline (y=100, color='black', linestyle='dotted')
plt.xlabel('Number of bets')
plt.ylabel('Winnings')
plt.title("Roulette Simulation")
```
